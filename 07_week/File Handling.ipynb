### Apply file mode 

```python
# r = read only
with open("./abc.txt", mode='r') as file:
    a = file.readlines()
    print(a)
```

```python
# w = write
with open("./abc.txt", mode='w') as file:
    a = file.readlines() # read file here
    print(a)
```

```python
# w = write
with open("./abc.txt", mode='r') as file:
    file.write("aaaaaaaaaaaaaaaa")
```

```python
# r+ = read and write
with open("./abc.txt", mode='r+') as file:
    a = file.readlines() # read file here
    print("before====",a)
    file.write("\nPakistan zinda bad!")
    a = file.readlines() # read file here
    print("After====",a)
```

```python
# r+ = read and write
with open("./abc.txt", mode='r+') as file:
    a = file.readlines() # read file here
    print(a)
    file.write("\nHello world!")
    file.seek(0)
    a = file.readlines() # read file here
    print(a)
```

```python
# r+ = read and write
with open("./abc.txt", mode='r+') as file:
    file.write("\nHello world!")
    file.seek(0)
    a = file.readlines() # read file here
    print(a)
```
